```python
# Пояснення: Щоб порахувати суму вектора a та транспонованого вектора b, треба виконати елементарне
# матричне множення. Однак, враховуючи, що a є вектором-рядком (розмірність (1, 5)), а b -
# вектором-стовпцем (розмірність (5, 1)), ми повинні використовувати правило матричного
# множення для отримання суми.  сума=[1×1/2+2×1+3×2+4×3+5×4]. Результат це скалярний добуток
# першого рядка вектора a на транспонований перший стовпець вектора b

import numpy as np
a = np.array([[1, 2, 3, 4, 5]])
b = np.array([[1/2, 1, 2, 3, 4]])

# Матричне множення
result = np.dot(a, b.T)

print("Сума вектора a та транспонованого вектора b:", result)
```

```python
import numpy as np

a = np.array([[1, 2, 3, 4, 5]])
b = np.array([[1/2, 1, 2, 3, 4]])

# Використовуємо матричний добуток
c = np.dot(a, b.T)

print("Матричний добуток вектора a та транспонованого вектора b:")
print(c)

# отримаємо матрицю c, яка є матричним добутком вектора a та
# транспонованого вектора b розмірністю (1, 1)
```

```python
# Пояснення: У матричному множенні важливо, щоб кількість стовпців
# першої матриці була рівна кількості рядків другої матриці. У нашому випадку,
# матриця a має розмірність (1, 5), а матриця b - (1, 5). При використанні
# np.dot(a, b), отримаємо помилку, оскільки кількість стовпців матриці
#  a (5) не дорівнює кількості рядків матриці b (1) для матричного множення.

import numpy as np

a = np.array([[1, 2, 3, 4, 5]])
b = np.array([[1/2, 1, 2, 3, 4]])

result = np.dot(a, b)

print(f"Матричний добуток вектора a та вектора b: {result}")
```

```python
# Пояснення: Добуток Адамара (Hadamard product) векторів визначається як поелементний
# добуток відповідних елементів двох векторів. У бібліотеці NumPy це можна виконати
# за допомогою оператора * Результат - новий вектор, у якому кожен елемент є добутком відповідних
# елементів векторів a та b.. result=[1×0.5, 2×1, 3×2, 4×3, 5×4]

import numpy as np
a = np.array([[1, 2, 3, 4, 5]])
b = np.array([[1/2, 1, 2, 3, 4]])

# добуток Адамара
result = a * b

print(f"Добуток Адамара вектора a та вектора b:  {result}")
```

```python
# Пояснення: ілення векторів визначається як поелементне ділення відповідних елементів
# двох векторів. У бібліотеці NumPy це можна виконати за допомогою оператора /.
# Результат- новий вектор, у якому кожен елемент є часткою відповідних елементів векторів a та b:
#result=[1÷0.5, 2÷1, 3÷2, 4÷3, 5÷4]

import numpy as np

a = np.array([[1, 2, 3, 4, 5]])
b = np.array([[1/2, 1, 2, 3, 4]])

# Поелементне ділення
result = a / b

print(f"Ділення вектора a на вектор b: {result}")
```

```python
# Пояснення: Для матричного ділення використовуємо оператор / або функцію np.divide -
# це виконає поелементне ділення між відповідними елементами вектора a та транспонованого вектора b

import numpy as np
a = np.array([[1, 2, 3, 4, 5]])
b = np.array([[1/2, 1, 2, 3, 4]])

# Поелементне ділення
result = np.divide(a, b.T)

print(f"Ділення вектора a на транспонований вектор b: \n{result}")
```

- Об'єднай (перемнож) попередні перетворення в одну матрицю та застосуй її до зображення

```python
import cv2 as cv
import urllib
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow as cv_imshow

# Read an image
def read_image_by_url(url):
    req = urllib.request.urlopen(url)
    arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
    img = cv.imdecode(arr, -1)
    return img

url1 = 'https://soklan.ru/list/pictures/77/ZLtKocskb1.jpg'
url2 = 'https://soklan.ru/list/pictures/77/4JWrcE1QeF.jpg'

img1 = read_image_by_url(url1)
img2 = read_image_by_url(url2)

# Відображення зображень у ноутбуці
plt.subplot(1, 2, 1)
plt.imshow(cv.cvtColor(img1, cv.COLOR_BGR2RGB))
plt.title('Image_1')
plt.subplot(1, 2, 2)
plt.imshow(cv.cvtColor(img2, cv.COLOR_BGR2RGB))
plt.title('Image_2')
plt.show()

# Зміна розміру другого зображення та додавання його в центр першого
rows, cols, _ = img1.shape

# Зміна розміру до чверті
img2_resized = cv.resize(img2, (cols // 4, rows // 4))

# Визначення координат для вставки img2_resized в центр img1
start_row = (rows - img2_resized.shape[0]) // 2
start_col = (cols - img2_resized.shape[1]) // 2
end_row = start_row + img2_resized.shape[0]
end_col = start_col + img2_resized.shape[1]

# Додавання img2_resized в центр img1
img1_with_img2 = img1.copy()
img1_with_img2[start_row:end_row, start_col:end_col] = img2_resized

# Відображення отриманого зображення
plt.imshow(cv.cvtColor(img1_with_img2, cv.COLOR_BGR2RGB))
plt.title('Image_1 with Image_2 added in the center')
plt.show()


# Афінні перетворення
# Зміщення першого зображення на 50 пікселів вправо та 50 пікселів вгору
translation_matrix = np.float32([[1, 0, 50], [0, 1, -50]])
img1_shifted = cv.warpAffine(img1, translation_matrix, (cols, rows))

# Поворот першого зображення на 42 градусів навколо центру та збільшення на 20%
rotation_angle = 42
scale_factor = 1.2
rotation_matrix = cv.getRotationMatrix2D((cols / 2, rows / 2), rotation_angle, scale_factor)
img1_rotated_and_scaled = cv.warpAffine(img1, rotation_matrix, (cols, rows))

# Відображення результатів афінних перетворень
plt.subplot(1, 2, 1)
plt.imshow(cv.cvtColor(img1_shifted, cv.COLOR_BGR2RGB))
plt.title('Image_1 Shifted')
plt.subplot(1, 2, 2)
plt.imshow(cv.cvtColor(img1_rotated_and_scaled, cv.COLOR_BGR2RGB))
plt.title('Image_1 Rotated and Scaled')
plt.show()


# Перетворення перспективи
# Задання матриць для різних перетворень
stretch_matrix = np.float32([[1.2, 0, 0], [0, 1.3, 0], [0, 0, 1]])
shift_matrix = np.float32([[1, 0, 50], [0, 1, 50], [0, 0, 1]])
shear_matrix = np.float32([[1, 0.05, 0], [0.1, 1, 0], [0, 0, 1]])
rotation_matrix = cv.getRotationMatrix2D((cols / 2, rows / 2), 10, 1)

# Об'єднання матриць у одну
combined_matrix = np.matmul(rotation_matrix, np.matmul(shear_matrix, np.matmul(shift_matrix, stretch_matrix)))

# Конвертація матриці в тип float32 та додавання останнього рядка
combined_matrix = np.vstack([combined_matrix, [0, 0, 1]]).astype(np.float32)

# Застосування комбінованої матриці до зображення
img1_perspective_transformed = cv.warpPerspective(img1, combined_matrix, (cols, rows))

# Відображення результатів перетворення перспективи
plt.imshow(cv.cvtColor(img1_perspective_transformed, cv.COLOR_BGR2RGB))
plt.title('Image_1 Perspective Transformed')
plt.show()
```
