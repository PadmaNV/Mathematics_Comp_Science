### ***Наближена залежність ефективності роботи:***

$$2\left(\frac{4}{1.2\sqrt{2\pi}}e^{-\frac{1}{2}\left(\frac{x-11}{1.2}\right)^2} +\frac{7}{2.4\sqrt{2\pi}}e^{-\frac{1}{2}\left(\frac{x-15}{2.4}\right)^2}\right)$$





# **Завдання 1**
Візуалізуй функцію на відрізку від 0 до 24. За допомогою SymPy порахуй невизначений інтеграл функції та інтеграл від a до b:

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, integrate, sqrt, pi, exp

# Задана функція
x = symbols('x')
f_x = 2 * ((4 / (1.2 * sqrt(2 * np.pi))) * exp(-0.5 * ((x - 11) / 1.2)**2) +
           (7 / (2.4 * sqrt(2 * np.pi))) * exp(-0.5 * ((x - 15) / 2.4)**2))

# Візуалізація функції
x_values = np.linspace(0, 24, 1000)
y_values = [f_x.subs(x, val) for val in x_values]

plt.plot(x_values, y_values, label='Функція')
plt.title('Графік функції')
plt.xlabel('Година')
plt.ylabel('Залежність ефективності')
plt.legend()
plt.show()

# Невизначений інтеграл
indefinite_integral = integrate(f_x, x)
print(f"Невизначений інтеграл: {indefinite_integral}")

# Інтеграл від a до b
a, b = 9, 18
definite_integral = integrate(f_x, (x, a, b))
print(f"Інтеграл від {a} до {b}: {definite_integral}")
```

# **Завдання 2**
Напиши функцію чисельного інтегрування методом прямокутників та порахуй інтеграл від a до b:

```python
def rectangular_integration(f, a, b, n):
    h = (b - a) / n
    result = 0
    for i in range(n):
        result += f(a + i * h)
    result *= h
    return result

result_rectangular = rectangular_integration(lambda x: 2 * ((4 / (1.2 * sqrt(2 * np.pi))) * exp(-0.5 * ((x - 11) / 1.2)**2) +
                                                           (7 / (2.4 * sqrt(2 * np.pi))) * exp(-0.5 * ((x - 15) / 2.4)**2)),
                                              a, b, 1000)
print(f"Чисельний інтеграл методом прямокутників: {result_rectangular}")
```

# **Завдання 3**
Напиши функцію чисельного інтегрування методом трапецій та порахуй інтеграл від a до b:

```python
def trapezoidal_integration(f, a, b, n):
    h = (b - a) / n
    result = 0.5 * (f(a) + f(b))
    for i in range(1, n):
        result += f(a + i * h)
    result *= h
    return result

result_trapezoidal = trapezoidal_integration(lambda x: 2 * ((4 / (1.2 * sqrt(2 * np.pi))) * exp(-0.5 * ((x - 11) / 1.2)**2) +
                                                            (7 / (2.4 * sqrt(2 * np.pi))) * exp(-0.5 * ((x - 15) / 2.4)**2)),
                                               a, b, 1000)
print(f"Чисельний інтеграл методом трапецій: {result_trapezoidal}")
```

# **Завдання 4**
Напиши функцію чисельного інтегрування методом Сімпсона та порахуй інтеграл від a до b:

```python
def simpsons_integration(f, a, b, n):
    h = (b - a) / n
    result = f(a) + f(b)
    for i in range(1, n, 2):
        result += 4 * f(a + i * h)
    for i in range(2, n-1, 2):
        result += 2 * f(a + i * h)
    result *= h / 3
    return result

result_simpsons = simpsons_integration(lambda x: 2 * ((4 / (1.2 * sqrt(2 * np.pi))) * exp(-0.5 * ((x - 11) / 1.2)**2) +
                                                      (7 / (2.4 * sqrt(2 * np.pi))) * exp(-0.5 * ((x - 15) / 2.4)**2)),
                                         a, b, 1000)
print(f"Чисельний інтеграл методом Сімпсона: {result_simpsons}")
```
