Використаєм формулу Бернуллі $P(X=k)=p^k * (1 - p)^{(n-k)}$:

де k - це кількість успішних результатів, n - загальна кількість спроб, p - ймовірність успіху в одній спробі. У нашому випадку n = 6, k = 2, і  p дорівнює відношенню кількості нестандартних деталей (5) до загальної кількості деталей (50): $$p = \frac{5}{50}$$
Тепер тепер підставляємо ці значення у формулу Бернуллі:
$$P(X=2) =\left( \frac{5}{50}\right)^2 * \left( 1 - \frac{5}{50}\right)^{6-2}$$  

Результат виразу - ймовірність того, що серед обраних шести деталей дві виявляться нестандартними.

Із 15 рейсів, що виконуються з аеропорту протягом доби, 60 % рейсів виконуються на власному літаковому парку. Знайди ймовірність того, що з вибраних навмання 5 рейсів рівно 3 виконуються на власному парку.

а) Розв’яжи аналітично.

б) Розв’яжи програмно, згенерувавши всі можливі комбінації. *Курсив*

## **2.a) Розв'язання аналітично:**
Використаєм формулу Бернуллі $P(X=k)=p^k * (1 - p)^{(n-k)}$. де
n - кількість спроб (рейсів), k - кількість успішних результатів (рейсів на власному літаку), p - ймовірність успіху в одній спробі, 1-p - ймовірність невдачі в одній спробі. У нашому випадку n=5, k=3, p=0,6:
$$P(X=3)=C_5^3 * (0.6)^3 * (1-0.6)^{6-2}$$
Розрахунок:

$$P(X=3)=\frac{5!}{3!*2!} * (0.6)^3 * (0.4)^2$$  
$$P(X=3)=10 * 0.216 * 0.16$$


$$P(X=3)=0.3456$$

Отже, ймовірність того, що з вибраних навмання 5 рейсів рівно 3 виконуються на власному парку, аналітично дорівнює 0.3456.

## **2.b) Розв'язання програмно:**

```python
from math import comb

def calculate_probability_analytical(n, k, p):
    q = 1 - p
    return comb(n, k) * (p**k) * (q**(n - k))

# Вхідні дані
total_flights = 5
success_flights = 3
probability_success = 0.6

# Розрахунок ймовірності за допомогою аналітичної формули
result_analytical = calculate_probability_analytical(total_flights, success_flights, probability_success)
print(f'Ймовірність : {result_analytical}. З вибраних навмання 5 рейсів рівно 3 виконуються на власному парку')
```

# **Завдання 3**

*Підкидають три гральні кубики. Що ймовірніше: отримати в сумі очок, що випали, 11 або 12? Розв’яжи програмно, згенерувавши всі можливі комбінації.*



```python
from itertools import product

def calculate_probabilities():
    outcomes_11 = 0
    outcomes_12 = 0
    total_outcomes = 0

    # Генерація всіх можливих комбінацій підкидання трьох кубиків
    for roll in product(range(1, 7), repeat=3):
        total_outcomes += 1

        # Рахуємо суму очок
        total_points = sum(roll)

        # Перевіряємо, чи випала сума 11 або 12
        if total_points == 11:
            outcomes_11 += 1
        elif total_points == 12:
            outcomes_12 += 1

    # Розрахунок ймовірностей
    probability_11 = outcomes_11 / total_outcomes
    probability_12 = outcomes_12 / total_outcomes

    return probability_11, probability_12

# Розрахунок ймовірностей
probability_11, probability_12 = calculate_probabilities()

# Виведення результатів
print(f'Ймовірність отримати суму очок 11: {probability_11:.4f}')
print(f'Ймовірність отримати суму очок 12: {probability_12:.4f}')
```

$P=\frac {A_{кільця}} {A_{велике  коло}} $

$P=\frac {\pi R^2_{більша}- \pi R^2_{меньша}}{\pi R^2_{меньша}} $

# **Завдання 6**

Урна містить одну кулю, про яку відомо, що вона або біла, або чорна з однаковими ймовірностями. В урну кладуть білу кулю і потім навмання виймають одну кулю. Вона виявилася білою. Яка ймовірність того, що куля, яка залишилася, є білою?

а) Розв’яжи аналітично.

б) Розв’яжи програмно за допомогою симуляції цього процесу достатньою кількістю разів.

## **6.a) Розв'язання аналітично:**

Позначимо події:
- A: витягнута куля є білою
- B: куля, яка залишилася, є білою

Ми шукаємо ймовірність події B при умові, що подія A відбулася.

Формула Байєса виглядає так:

$ P(B|A) = \frac{P(A|B) \cdot P(B)}{P(A)} $

де:
- $ P(B|A) $ - ймовірність того, що куля, яка залишилася, є білою при умові, що витягнута куля біла.
- $ P(A|B) $ - ймовірність того, що витягнута куля є білою при умові, що залишилася біла куля.
- $ P(B) $ - апріорна ймовірність того, що куля, яка залишилася, є білою (без урахування витягнутої кулі).
- $ P(A) $ - апріорна ймовірність того, що витягнута куля є білою.

За умови, що куля, яка залишилася, є білою (подія B відбулася), апріорна ймовірність цього становить 1/2 (оскільки перед витягненням кулі в урні була рівна кількість білих і чорних куль). Тобто,  $P(B) = \frac{1}{2} $.

Ймовірність того, що витягнута куля є білою при умові, що залишилася біла куля (подія A при умові B), також дорівнює 1, оскільки в нас вже відомо, що залишилася біла куля. Тобто, $ P(A|B) = 1 $.

Ймовірність того, що витягнута куля є білою (подія A) дорівнює ймовірності того, що вона біла і перед витягненням не була залишена в урні (тобто чорна була залишена), плюс ймовірність того, що вона біла і перед витягненням була залишена в урні (тобто біла залишена) і дорівнює $ P(A) = \frac{1}{2} \cdot 0 + \frac{1}{2} \cdot 1 = \frac{1}{2} $.

Тепер можемо використовувати Формулу Байєса:

$ P(B|A) = \frac{1 \cdot \frac{1}{2}}{\frac{1}{2}} = 1 $

Отже, ймовірність того, що куля, яка залишилася, є білою, дорівнює 1.

# **6.b) Розв'язання програмно:**

```python
import random

# Функція для симуляції витягнення куль з урни
def simulate():
    global count_all, count_second_white
    box = ["B", "W"]
    # Витягнути першу кулю
    first = box.pop()

    if first == "W":
        # Якщо витягнута куля біла, залишити її в урні
        count_all += 1

        # Витягнути другу кулю
        second = box.pop()

        if second == "W":
            # Якщо друга куля теж біла, збільшити лічильник білих куль
            count_second_white += 1

count_all = 0
count_second_white = 0

# Симуляція
num_simulations = 100000
for _ in range(num_simulations):
    simulate()

# Обчислити ймовірність того, що куля, яка залишилася, є білою
probability = count_second_white / count_all

print("Ймовірність (програмно):", probability)

```

# **Завдання 7. Випадкове блукання**

Ціна акцій компанії в кожен момент часу може з рівною ймовірністю збільшитись на 2 або зменшитись на 1. У початковий момент часу t=0 ціна рівна 0. Визнач середню ціну акції через 3 одиниці часу (t=3).
а) Розв’яжи аналітично.
б) Напиши симуляцію даного процесу. Порахуй середнє значення та намалюй гістограму ціни для 10, 100, 1000 та 10000 симуляцій:

## **7.a) Розв'язання аналітично:**
У кожен момент часу ціна може збільшитись на 2 або зменшитись на 1.

Запишемо можливі значення ціни після трьох одиниць часу:


1. Ціна може збільшитись на 2 одиниці (2 рази).
2. Ціна може збільшитись на 2 одиниці та потім зменшитись на 1 одиницю (1 раз).
3. Ціна може збільшитись на 2 одиниці та потім знову збільшитись на 2 одиниці (1 раз).
4. Ціна може зменшитись на 1 одиницю та потім збільшитись на 2 одиниці (1 раз).
5. Ціна може зменшитись на 1 одиницю та потім знову зменшитись на 1 одиницю (1 раз).

Розрахуємо ймовірність кожної з цих подій та визначимо очікуване значення:

$E[X]=∑_i​P_i​ ⋅X_i​ $

де $ X$ - ціна акцій, $P_i​ $ - ймовірність події $X_i$.

$E [X]=2 * \frac {2}{4} +1 *\frac {1}{4} +4 *\frac {1}{4}-1 *\frac {1}{4}$

Отже, очікуване значення ціни акцій через 3 одиниці часу дорівнює $\frac {3}{2}$ .

## **7.b) Симуляція та гістограма:**

```python
import random
import matplotlib.pyplot as plt

def simulate_stock_price(num_simulations, time_steps):
    results = []

    for _ in range(num_simulations):
        price = 0

        for _ in range(time_steps):
            # Генеруємо випадковий крок (збільшення на 2 або зменшення на 1)
            step = random.choice([2, -1])
            price += step

        results.append(price)

    return results

def plot_histogram(data, bins, title):
    plt.hist(data, bins=bins, edgecolor='black')
    plt.title(title)
    plt.xlabel('Ціна акцій')
    plt.ylabel('Кількість симуляцій')
    plt.show()

# Проведемо симуляції для різної кількості разів
num_simulations_list = [10, 100, 1000, 10000]
time_steps = 3

for num_simulations in num_simulations_list:
    simulated_prices = simulate_stock_price(num_simulations, time_steps)
    plot_histogram(simulated_prices, bins=20, title=f'Гістограма ціни акцій для {num_simulations} симуляцій')
```
