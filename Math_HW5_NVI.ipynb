## **Завдання 1**
**Виконай диференціювання функції та знайди значення похідних у точках x=1 і x=−1/2. Розв’яжи програмно та аналітично.**

Завдання 1.jpg

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, sqrt

# Визначення символьних змінних
x = symbols('x')

# Визначення функцій
f1 = x**3/3 + x**2/2 - 2*x
f2 = sqrt(x**2 + 1)
f3 = 1/sqrt(x**2 + 1)

# Знайдення похідних
f1_prime = diff(f1, x)
f2_prime = diff(f2, x)
f3_prime = diff(f3, x)

# Визначення значень похідних у точках x=1 і x=-1/2
x_values = [1, -1/2]
f1_prime_values = [f1_prime.subs(x, val) for val in x_values]
f2_prime_values = [f2_prime.subs(x, val) for val in x_values]
f3_prime_values = [f3_prime.subs(x, val) for val in x_values]

# Виведення результатів
print("f1'(1) =", f1_prime_values[0])
print("f1'(-1/2) =", f1_prime_values[1])
print("f2'(1) =", f2_prime_values[0])
print("f2'(-1/2) =", f2_prime_values[1])
print("f3'(1) =", f3_prime_values[0])
print("f3'(-1/2) =", f3_prime_values[1])

# Побудова графіків
x_vals = np.linspace(-2, 2, 400)
f1_vals = [f1.subs(x, val) for val in x_vals]
f2_vals = [f2.subs(x, val) for val in x_vals]
f3_vals = [f3.subs(x, val) for val in x_vals]
f1_prime_vals = [f1_prime.subs(x, val) for val in x_vals]
f2_prime_vals = [f2_prime.subs(x, val) for val in x_vals]
f3_prime_vals = [f3_prime.subs(x, val) for val in x_vals]

# Побудова графіків
plt.figure(figsize=(12, 8))

# Графік функцій
plt.subplot(2, 1, 1)
plt.plot(x_vals, f1_vals, label='f1(x)')
plt.plot(x_vals, f2_vals, label='f2(x)')
plt.plot(x_vals, f3_vals, label='f3(x)')
plt.legend()
plt.title('Графік функцій')

# Графік похідних
plt.subplot(2, 1, 2)
plt.plot(x_vals, f1_prime_vals, label="f1'(x)")
plt.plot(x_vals, f2_prime_vals, label="f2'(x)")
plt.plot(x_vals, f3_prime_vals, label="f3'(x)")

# Маркування точок
plt.scatter(x_values, f1_prime_values, color='red')
plt.scatter(x_values, f2_prime_values, color='red')
plt.scatter(x_values, f3_prime_values, color='red')
plt.legend()
plt.title('Графік похідних')

plt.show()
```

## **Завдання 2**
**Виконай диференціювання функції та знайди точку максимуму. Візуалізуй функцію, її похідну та точку максимуму. Розв’яжи лише програмно.**

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, solve

# Визначення символьної змінної
x = symbols('x')

# Визначення функції
y = -3*x**2 + 30*x

# Знайдення похідної
y_prime = diff(y, x)

# Знайдення точок, де похідна дорівнює нулю (критичні точки)
critical_points = solve(y_prime, x)

# Знаходимо значення функції в критичних точках та обираємо точку максимуму
maximum_point = (critical_points[0], y.subs(x, critical_points[0]))

# Побудова графіків
x_vals = np.linspace(0, 10, 400)
y_vals = [-3*x**2 + 30*x for x in x_vals]
y_prime_vals = [-6*x + 30 for x in x_vals]

plt.figure(figsize=(10, 6))

# Графік функції
plt.plot(x_vals, y_vals, label='y = -3x^2 + 30x')

# Графік похідної
plt.plot(x_vals, y_prime_vals, label="y' = -6x + 30")

# Маркування точки максимуму
plt.scatter(maximum_point[0], maximum_point[1], color='red', label=f'Maximum Point {maximum_point}')

plt.legend()
plt.title('Графік функції та її похідної')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
