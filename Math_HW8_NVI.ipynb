## **Завдання 1:**
а) Побудуй гістограму розподілу x.

б) Запусти симуляцію з t від 1 до, наприклад, ~60 з кроком, наприклад, 1 або 2.

```python
import random
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma, normaltest


def simulate_stock_price(num_simulations, time_steps):
    results = []

    for _ in range(num_simulations):
        price = 0

        for _ in range(time_steps):
            # Генеруємо крок за гамма-розподілом
            step = int(np.random.gamma(0.3, 1.1))
            price += step

        results.append(price)

    return results

def plot_histogram(data, bins, title):
    plt.hist(data, bins=bins, edgecolor='black')
    plt.title(title)
    plt.xlabel('Ціна акцій')
    plt.ylabel('Кількість симуляцій')
    plt.show()

# а) Побудова гістограми розподілу x
num_simulations = 100
time_steps = 3
simulated_prices = simulate_stock_price(num_simulations, time_steps)
plot_histogram(simulated_prices, bins=20, title='Гістограма розподілу x')

# б) Побудова гістограм та перевірка на нормальність для різних значень t
for time_steps in range(1, 61, 2):
    simulated_prices = simulate_stock_price(num_simulations, time_steps)
    plot_histogram(simulated_prices, bins=20, title=f'Гістограма ціни акцій для {time_steps} симуляцій')

    # Перевірка на нормальність
    stat, p_value = normaltest(simulated_prices)
    if p_value < 0.05:
        print(f'Розподіл для {time_steps} симуляцій не є нормальним (p-value: {p_value})')
    else:
        print(f'Розподіл для {time_steps} симуляцій є нормальним (p-value: {p_value})')
```

**На основі проведених симуляцій та тестів на нормальність можна зробити наступні ВИСНОВКИ:**

Гістограма розподілу x (при фіксованому t=3):

1.   *Гістограма розподілу x, отримана при фіксованому значенні t=3, показує, що розподіл не є схожим на нормальний. Він має складну форму з вираженими відхиленнями та деякими піковими значеннями.*
2.   *Зміна розподілу зі збільшенням t:
При збільшенні значення t (кількості кроків в симуляції), можна помітити, що форма розподілу розширюється та стає більш симетричною. Це вказує на те, що зі збільшенням кількості кроків розподіл стає більш "гладким" та наближеним до нормального. **
3.   *Тест на нормальність:
Результати тестів на нормальність (використовуючи тест Д'Агостино) підтверджують ці спостереження. При менших значеннях t, розподіл не є нормальним (p-value < 0.05), але зі збільшенням t p-value може збільшуватись, що вказує на підвищення нормальності розподілу. **

Отже, можна висловити припущення, що зі збільшенням кількості кроків у симуляції розподіл ціни акцій наближається до нормального розподілу, що відображає класичну статистичну теорію (центральна гранична теорема), за якою сума великої кількості незалежних випадкових змінних має нормальний розподіл.
